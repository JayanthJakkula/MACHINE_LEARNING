```python
import pandas as pd
import numpy as np
df=pd.read_csv("heart (1).csv")
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoders={}
for col in df.select_dtypes(include=['object']).columns:
    le=LabelEncoder()
    df[col]=le.fit_transform(df[col])
    label_encoders[col]=le
    
```

```python
x=df.drop(columns=['target'])
y=df['target']
# split data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
# train with svm
from sklearn.svm import SVC
svm_linear=SVC(kernel='linear',C=1)
svm_linear.fit(x_train,y_train)
y_pred_linear=svm_linear.predict(x_test)
```

```python
from sklearn.metrics import *
print("confusion matrix :",confusion_matrix(y_test,y_pred_linear))
print("Accuracy :",accuracy_score(y_test,y_pred_linear))
```

```python
# train svm with poly kernel
svm_poly=SVC(kernel='poly',degree=3,C=1)
svm_poly.fit(x_train,y_train)
y_pred_poly=svm_poly.predict(x_test)
```

```python
print("confusion matrix :",confusion_matrix(y_test,y_pred_poly))
print("Accuracy :",accuracy_score(y_test,y_pred_poly))
```

```python
# train svm with rbf kernel
svm_rbf=SVC(kernel='rbf',C=1,gamma='scale')
svm_rbf.fit(x_train,y_train)
y_pred_rbf=svm_rbf.predict(x_test)
print("confusion matrix :",confusion_matrix(y_test,y_pred_rbf))
print("Accuracy :",accuracy_score(y_test,y_pred_rbf))
```
